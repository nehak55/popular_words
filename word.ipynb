```python
def get_popular_words(directory):
    
    """
    This function will return most popular words in the documents, the document name and the sentences 
    where the words are used
    """
    
    
    df = pd.DataFrame([]) # stores the final result
    
    for docs in os.listdir(directory):
        fname = os.path.join(directory, docs)
        if os.path.isfile(fname):
            with open(fname) as f:
                
                text_string = f.read().lower()
                
                # Check words of minimum 4 characters
                match_pattern = (re.findall(r'\b[a-z]{4,19}\b', text_string)) 
                
                # Get rid of stop words
                filtered_words = [fw for fw in match_pattern if fw not in stopwords.words('english')]
                #ps = PorterStemmer()
                #filtered_words = [ps.stem(i) for i in filtered_words]
            
                
                fdist = nltk.FreqDist(filtered_words)
                top_freq = fdist.most_common(5)
                top_five = [w for w,f in top_freq] # Get top 5 most popular words in a document
                
                
                # Check for sentences where these popular words appeared
                sent_tokenize = nltk.sent_tokenize(text_string) 
                for sentence in sent_tokenize:
                    token = re.findall(r'\b[a-z]{4,19}\b', sentence)
                    #match_stemmer = [ps.stem(i) for i in match]
                    for word in top_five:
                        if word in token:
                            df = df.append(pd.DataFrame({'Docs' : docs, 'word' : word, 'sentences' : sentence},
                                                   index=[0]), ignore_index=True)
  
                            
    df= df.groupby(['word','Docs'])['sentences'].apply(' '.join).reset_index()
    return(df.style.set_properties(subset=['sentences'], **{'width': '500px'}))
    
```
